<h5 Style='color:#00af83'>4.1 Identificar valores faltantes</h5>

```python
dfAutos.replace('?',np.nan,inplace=True)# reemplaza los valores faltantes por el estandar que es NaN
dfAutos.head(5)
```

<h5 Style='color:#00af83'>4.2 Contar la cantidad de valores faltantes</h5>

```python
dfTemp=dfAutos.isnull()# dataframe temporal, se va al dataframe autos y donde hay un campo nulo se va a poner un TRUE 
# y cuando no esta nulo se pone un false
dfTemp.head(5)
```

```python
for column in dfTemp.columns.values.tolist(): # convierta los valores del dataframe temp y cree una lista q se almacena en column
    print(column)
    print(dfTemp[column].value_counts())
    print("") # imprime los valores faltantes por cada columna
```

<h5 Style='color:#00af83'>4.3 Imputacion de valores numericos</h5>

```python
#Hay datos que no se pueden emputar porque puede haber una implicacion respecto a la toma de decisiones
av_normalized=dfAutos["normalized"].astype("float").mean(axis=0)# vaya al dataframe a la columna de normalized conviertela
# momentaneamente en un tipo de datos numericos y calcule la media de todos los datos de esa columna 
print('La media es:', av_normalized)
```

```python
dfAutos["normalized"].replace(np.nan,122,inplace=True)
dfAutos.head(5)
```

```python
av_normalized=dfAutos["normalized"].astype("float").mean(axis=0)
av_bore=dfAutos["bore"].astype("float").mean(axis=0)
av_stroke=dfAutos["stroke"].astype("float").mean(axis=0)
av_horsepower=dfAutos["horsepower"].astype("float").mean(axis=0)
av_peak=dfAutos["peak"].astype("float").mean(axis=0)
```

```python
dfAutos["normalized"].replace(np.nan,av_normalized,inplace=True)
dfAutos["bore"].replace(np.nan,av_bore,inplace=True)
dfAutos["stroke"].replace(np.nan,av_stroke,inplace=True)
dfAutos["horsepower"].replace(np.nan,av_horsepower,inplace=True)
dfAutos["peak"].replace(np.nan,av_peak,inplace=True)
dfAutos.head(5)
```

<h5 Style='color:#bb486d'>5.2 Convertir valores numericos en funcion del medio (Estandarizar) </h5>

<h5 Style='color:#00af83'>5.3 Normalizar valores entre 0 y 1 </h5>

```python
dfTemp = dfAutos[["City L/100km (Calc)", 'Higway L/100km (Calc)',"engineSize"]]
dfTemp.head(5)
```

```python
dfTemp.describe()
```

```python
dfAutos["City L/100km (norm)"]= dfAutos['City L/100km (Calc)']/dfAutos['City L/100km (Calc)'].max() 
dfAutos["Higway L/100km (norm)"]= dfAutos['Higway L/100km (Calc)']/dfAutos['Higway L/100km (Calc)'].max() 
dfAutos["engineSize(norm)"]= dfAutos['engineSize']/dfAutos['engineSize'].max()
dfTemp = dfAutos[["City L/100km (Calc)", 'Higway L/100km (Calc)',"engineSize", "City L/100km (norm)",
                  "Higway L/100km (norm)","engineSize(norm)"]]
dfTemp.head(5) #normalizo los valores entre 0 y 1
# no es necesario usarlo en el puntero
# sirve para comparar variables que tienen rangos muy distintos entre si
# se ve como valor absoluto y no se puede ver como porcentaje
```

```python
import matplotlib as mpl #Para configurar las graficas.
#Remove all axis, sirve para que no salgan las lineas solo la de abajo
mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = False
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
gf = dfAutos.groupby(['make']).count()

gf = gf.iloc[:,:1]

gf = gf.sort_values('symboling', ascending = False)

gf

fig, ax = plt.subplots(1,1, figsize=(15,6.9), dpi=100)

eje_x = gf.index.tolist()

data = gf['symboling'].tolist()

bar_colors = ['#89abb4','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',

              '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',

              '#e5e5e5','#fdd1b6']

ax.barh(eje_x, data, color=bar_colors)


#--------------------------------------------------------------------------------------------------------------------#

#--------------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EJES DEL GRAFICO ------------------------------------#

#--------------------------------------------------------------------------------------------------------------------#

#FORMATO DEL TITULO

title = "Grafico 1: Vehiculos"

plt.suptitle(title,

             horizontalalignment = 'left',

             x = 0.001,

             y = 0.87,

             transform = fig.transFigure,

             fontsize = 20,

             fontweight = "bold",

             fontfamily = "serif")


#FORMATO DEL SUBTITULO

subTitle = "Cantidad de vehiculos agrupados segun fabricante"

ax.set_title(subTitle,

             horizontalalignment = 'left',

             x = 0.001,

             y = 0.80,

             transform = fig.transFigure,

             fontsize = 15,

             fontweight = "regular",

             fontfamily = "serif")


#FORMATO DE LOS EJES

etiqueta = "Cantidad de vehiculos (ABS)"

ax.set_xlabel(etiqueta,

              ha = 'left',

              fontsize = 10,

              fontweight = "regular",

              fontfamily = "serif")

#CONFIGURACION DE COORDENADAS DE LA ETIQUETA DEL EJE X

ax.xaxis.set_label_coords(x=0,

                          y=-0.07,

                          transform=ax.transAxes)

#CONFIGURACION EL TAMANO DEL TEXTO DEL EJE Y

ax.tick_params(axis='both',

               labelsize=10)

#CONFIGURACION LA UBICACION DEL GRAFICO

plt.subplots_adjust(top=0.80)

plt.show()
```
